# Encoding in Regression Model

Label Encoding 

One-Hot Encoding


Types about variable

```python
import pandas as pd
```

```python
df = pd.read_csv("Car Model.csv")
```

```python
dummies = pd.get_dummies(df['Car Model'])
```

```python
dummies 
```

```python
merged = pd.concat([df,dummies],axis='columns')
```

```python
merged
```

```python
merged2 = merged.drop(['Car Model'],axis='columns')
```

```python
merged2
```

```python
# Output is y
y= df['Sell Price']
```

```python
y
```

```python
# input is x
x= merged2.drop(['Sell Price'],axis=1)
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression

```

```python
model= LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[784554,6,0,1,0]])
```
