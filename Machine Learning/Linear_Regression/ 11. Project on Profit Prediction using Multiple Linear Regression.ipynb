# Separate x,y

```python
x = df.drop(['Profit'],axis=1)
```

```python
x
```

```python
y = df['Profit']
```

```python
y
```

# One hot coding

```python
# Convert the column into categorical columns

city = pd.get_dummies(x['Area'],drop_first=True)
```

```python
city
```

```python
# Drop the Area coulam
x  =  x.drop('Area',axis=1)
```

```python
x.head()
```

```python
# Now concatation
x= pd.concat([x,city],axis=1)
```

```python
x
```

```python
# Spliting the dataset into the Training set and Test set
```

```python
# import library
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# create linear regression 
regessor = LinearRegression()
```

```python
regessor.fit(xtrain,ytrain)
```

```python
ytest
```

```python
# Predict input is feature
xtest
```

```python
# predicitng the Test set results
pred = regessor.predict(xtest)
```

```python
pred
```

```python
# check Performance 

regessor.score(xtest,ytest)
```

# R- Squared Value 

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(ytest,pred)
```
